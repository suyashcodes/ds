```python
# %% [markdown]
# # Descriptive Statistics - Measures of Central Tendency and Variability
# 
# This notebook demonstrates descriptive statistics operations on two different datasets.

# %%
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# Set style for plots
sns.set_style('whitegrid')

# %% [markdown]
# ## Part 1: Summary Statistics Grouped by Categorical Variable

# %%
# Load a sample dataset (we'll use the built-in tips dataset from seaborn)
tips = sns.load_dataset('tips')
print("Dataset Overview:")
display(tips.head())

# %%
# 1. Provide summary statistics for numeric variables grouped by a categorical variable
# Let's analyze 'total_bill' and 'tip' grouped by 'time' (Lunch/Dinner)

# Select numeric columns
numeric_cols = ['total_bill', 'tip', 'size']

# Group by the categorical variable 'time' and calculate statistics
grouped_stats = tips.groupby('time')[numeric_cols].agg(
    ['mean', 'median', 'min', 'max', 'std', 'count']
)

print("\nSummary Statistics Grouped by Time (Lunch/Dinner):")
display(grouped_stats)

# %%
# Create a list containing numeric values for each response to the categorical variable
# For example, list of all total_bill amounts for Lunch and Dinner
lunch_bills = tips[tips['time'] == 'Lunch']['total_bill'].tolist()
dinner_bills = tips[tips['time'] == 'Dinner']['total_bill'].tolist()

print("\nList of total_bill values for Lunch:", lunch_bills[:5], "...")
print("List of total_bill values for Dinner:", dinner_bills[:5], "...")

# %%
# Visual representation of the grouped statistics
plt.figure(figsize=(12, 6))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(1, 3, i)
    sns.boxplot(x='time', y=col, data=tips)
    plt.title(f'{col} by Time')
plt.tight_layout()
plt.show()

# %% [markdown]
# ## Part 2: Basic Statistical Details of Iris Species

# %%
# Load the iris dataset
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target_names[iris.target]

print("\nIris Dataset Overview:")
display(iris_df.head())

# %%
# 2. Display basic statistical details for each iris species
# Select the numeric features
iris_numeric = iris_df.columns[:-1]

# Group by species and calculate statistics
iris_stats = iris_df.groupby('species')[iris_numeric].agg(
    ['mean', 'median', 'std', 'min', 'max', 
     lambda x: np.percentile(x, 25),  # 25th percentile
     lambda x: np.percentile(x, 75)   # 75th percentile
])

# Rename the percentile columns
iris_stats.columns = [' '.join(col).strip() for col in iris_stats.columns.values]
iris_stats = iris_stats.rename(columns={
    '<lambda_0>': '25th percentile',
    '<lambda_1>': '75th percentile'
})

print("\nDetailed Statistics by Iris Species:")
display(iris_stats)

# %%
# Alternative way using describe() for percentiles
print("\nDetailed Statistics using describe():")
display(iris_df.groupby('species')[iris_numeric].describe().transpose())

# %%
# Visual comparison of the species
plt.figure(figsize=(15, 10))
for i, feature in enumerate(iris_numeric, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(x='species', y=feature, data=iris_df)
    plt.title(f'{feature} by Species')
plt.tight_layout()
plt.show()

# %%
# Pairplot to visualize relationships between variables
sns.pairplot(iris_df, hue='species', diag_kind='hist')
plt.suptitle('Pairwise Relationships by Iris Species', y=1.02)
plt.show()
```
