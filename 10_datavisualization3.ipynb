```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Load the Iris dataset
iris = sns.load_dataset('iris')

# View first few rows
print(iris.head())
# Features and Data Types
print(iris.dtypes)
# Plot histograms for all numeric features
iris.hist(figsize=(10, 8), edgecolor='black')
plt.suptitle('Histograms of Iris Dataset Features', fontsize=16)
plt.show()
# Plot boxplots for all numeric features
plt.figure(figsize=(12,8))

for idx, feature in enumerate(iris.columns[:-1]):
    plt.subplot(2, 2, idx+1)
    sns.boxplot(y=iris[feature])
    plt.title(f'Boxplot of {feature}')

plt.tight_layout()
plt.show()
```
